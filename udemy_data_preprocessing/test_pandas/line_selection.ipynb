query: 行の抽出

行の抽出では基本的にqueryを使用したほうがgood

```python
import pandas as pd
import numpy as np
import seaborn as sns

pd.set_option('display.max_rows', 5)
```

```python
df = sns.load_dataset('diamonds')
```

```python
df
```

bad: price列が337の行を抽出

```python
#論理値ベクトルを出力
df["price"] == 337
```

```python
#論理値ベクトルのTRUEのみを出力
df[df["price"] == 337]
```

good: price列が337の行を抽出 queryを使用

```python
df.query("price == 337")
```

depth列が62以上の行を抽出

```python
df.query("depth >= 62")
```

carat列が0.23~0.27の行を抽出

```python
df.query("0.23 <= carat <= 0.27")
```

color列がEの行を抽出<br>
文字列の場合はシングルorダブルクオーテーションで囲む<br>
この時、同じ種類のクオーテーションは使用しない

```python
df.query("color == 'E'")
```

color列がE以外の列を抽出

```python
df.query("color != 'E'")
```

color列がIまたはJの列を抽出

```python
df.query("color in['I', 'J']")
```

bad: depth列が62以上かつcolor列がHの行を抽出（AND）

```python
(df["depth"] >= 62) & (df["color"] == 'E')
```

```python
df[(df["depth"] >= 62) & (df["color"] == 'E')]
```

good：queryを使用

```python
df.query("depth >= 62 & color =='E'")
```

depth列が62以上またはcolor列がHの行を抽出(OR)

```python
df.query("depth >= 62 | color =='H'")
```

（depth列が62以上かつcolor列がH）でない列を抽出(NOT)

```python
df.query("~(depth >= 62 & color =='H')")
```

price列が337の行を抽出（変数を使用）<br>
変数の前に@

```python
x = 337
df.query("price == @x")
```
