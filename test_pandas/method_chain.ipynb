メソッドチェーン こっちが可読性がよいコード

```python
import numpy as np
import pandas as pd
import seaborn as sns

#データフレームの表示行数を指定
pd.set_option('display.max_rows', 5)
```

```python
#データ読み込み
df = sns.load_dataset('iris')
```

```python
df
```

```python
#列の更新
#sepal_length + sepal_widthをしてsepal_lengthに代入
df.assign(sepal_length = lambda x: x["sepal_length"] + x["sepal_width"])
```

```python
#行の抽出
df.assign(sepal_length = lambda x: x["sepal_length"] + x["sepal_width"]).query("sepal_length > 8")
```
