ソート

```python
import numpy as np
import pandas as pd
import seaborn as sns 

pd.set_option("display.max_rows", 5)
```

```python
df = sns.load_dataset("diamonds")
```

```python
df
```

depth列を昇順にソート

```python
df.sort_values("depth", ascending=True)
```

depth列を昇順にソート(ascending = True省略可)

```python
df.sort_values("depth")
```

depth列を降順にソート

```python
df.sort_values("depth", ascending=False)
```

複数列(depth列、table列)でソート

```python
df.sort_values(["depth", "table"])
```
